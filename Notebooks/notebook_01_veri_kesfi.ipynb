```python
import pandas as pd
import numpy as np

file_path = 'elektrik_veri_hashed.xlsx'
xlsx = pd.ExcelFile(file_path)

print("Sayfa isimleri:", xlsx.sheet_names)

df_tahsilat = pd.read_excel(xlsx, sheet_name='Tahsilat')
df_tahsilat_1 = pd.read_excel(xlsx, sheet_name='Tahsilat 1')
df_tahakkuk = pd.read_excel(xlsx, sheet_name='Tahakkuk')
df_tahakkuk_1 = pd.read_excel(xlsx, sheet_name='Tahakkuk 1')
df_tahakkuk_2 = pd.read_excel(xlsx, sheet_name='Tahakkuk 2')

print(f"Tahsilat: {df_tahsilat.shape}")
print(f"Tahsilat 1: {df_tahsilat_1.shape}")
print(f"Tahakkuk (Hamamözü): {df_tahakkuk.shape}")
print(f"Tahakkuk 1 (Gümüşhacıköy): {df_tahakkuk_1.shape}")
print(f"Tahakkuk 2 (Göynücek): {df_tahakkuk_2.shape}")
```

```python
tahakkuk_dfleri = {
    'Hamamözü': df_tahakkuk,
    'Gümüşhacıköy': df_tahakkuk_1,
    'Göynücek': df_tahakkuk_2
}

for ilce_adi, df in tahakkuk_dfleri.items():
    print(f"\n{'*'*40}")
    print(f"{ilce_adi} Tahakkuk Verisi Özeti")
    print(f"{'*'*40}")

    print("\n--- INFO ---")
    df.info()

    print("\n--- DESCRIBE ---")
    display(df.describe())

    print("\n--- HEAD (İlk 3 Satır) ---")
    display(df.head(3))
```

```python
df_tahakkuk['ilce'] = 'Hamamözü'
df_tahakkuk_1['ilce'] = 'Gümüşhacıköy'
df_tahakkuk_2['ilce'] = 'Göynücek'

tum_tahakkuk = pd.concat([df_tahakkuk, df_tahakkuk_1, df_tahakkuk_2], ignore_index=True)

print(f"Birleştirilen verideki toplam kayıt sayısı: {len(tum_tahakkuk)}")

benzersiz_musteri = tum_tahakkuk.groupby('ilce')['sozlesme_hesap_no'].nunique().reset_index()
benzersiz_musteri.columns = ['İlçe', 'Benzersiz Müşteri Sayısı']

print("\n*** İlçelere Göre Benzersiz Müşteri Sayıları ***")
display(benzersiz_musteri)
```

```python
eksik_kwh = tum_tahakkuk['kwh'].isna().sum()
print(f"Eksik kwh değeri sayısı: {eksik_kwh}")

negatif_kwh = len(tum_tahakkuk[tum_tahakkuk['kwh'] < 0])
print(f"Negatif kwh değerine sahip kayıt sayısı: {negatif_kwh}")

Q1 = tum_tahakkuk['kwh'].quantile(0.25)
Q3 = tum_tahakkuk['kwh'].quantile(0.75)
IQR = Q3 - Q1

alt_sinir = Q1 - 1.5 * IQR
ust_sinir = Q3 + 1.5 * IQR

uc_degerler = tum_tahakkuk[(tum_tahakkuk['kwh'] < alt_sinir) | (tum_tahakkuk['kwh'] > ust_sinir)]
print(f"Aşırı uç kwh (outlier) kayıt sayısı: {len(uc_degerler)}")
```
